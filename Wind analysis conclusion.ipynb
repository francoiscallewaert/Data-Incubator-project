In this project, we aim to answer the question: “How much of the total French electricity consumption can wind energy provide?”. And we could add “In a reasonable manner”.

In order to answer the question, we plan to compare the hourly electricity consumption with the hourly electricity generation from wind to evaluate which proportion of the total consumption could be provided by wind power. In this early work, we assume a constant power consumption normalized to 1 and we study multiple plant configurations to optimize the amount of energy that can be provided by wind power. 

To answer that question, we rely on a recent dataset from the “European Meteorological derived High resolution” (EMHIRES) that provides hourly time series on 30 years of wind power potential in 255 regions in Europe, 22 of them in France. The data is also available on Kaggle (https://www.kaggle.com/sohier/30-years-of-european-wind-generation).<br />
This data is analyzed in a more detailed dataset at: https://github.com/francoiscallewaert/Data-Incubator-project/blob/master/Wind%20data%20analysis.ipynb <br />
Here we present the main conclusions.




We studied multiple wind plants configuration, where wind power is produced onshore in one region, or 2 regions (Normandy and Rhone), or when it is produced offshore, or with a mix onshore-offshore. For the first two cases, we considered both a plant with a maximum power corresponding to the global power consumption, and we also considered an oversized plant with 20% curtailment.

The first graph summarizes the percentage of the global electricity consumption that wind power currently provides, and could provide under the 6 cases described above.

```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
Groups = ('Current', '1region', '1region_oversized', '2regions', '2regions_oversized', 'offshore', 'offshore+onshore')
Energy = (4, 36, 50, 50, 65, 60, 72)
 
# create plot
fig = plt.figure(figsize = (14, 8))
index = np.arange(7)
bar_width = 0.75
opacity = 0.8
 
rects1 = plt.bar(index, Energy, bar_width,
                 alpha=opacity,
                 color='b')
 
plt.ylabel('Energy(%)', fontsize = 15)
plt.title('Percentage of total energy consumption from wind for various plant configurations', fontsize = 20)
plt.xticks(index, Groups, fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend()
 
plt.tight_layout()
plt.show()
```

Let's considered the '2regiond_oversized' configuration. We want to know, over 30 years, during how many hours will wind provide 100% of the consumption, how many hours will it be 0-10%, 40-50%, so that we can evaluate what kind of other electricity sources (hydro, solar, storage, gas, nuclear...) are needed to provide the remaining electricity to match demand.<br />
As can be seen, wind provides 100% of demand 36% of the time. This is nice, but wind provides less than 50% of demand 37% of the time, and less than 10% demand 8% of the time. This shows there are a lot of technical challenges to solve related to intermittent generation under a 60% wind scenario. We propose to study that more in detail
