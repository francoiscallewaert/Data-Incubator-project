```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
Groups = ('Current', 'Normandy', 'Normandy + Rhone', 'N+R + 6h storage', 'N+R + 24h storage', 'goal')
Energy = (4, 53, 70, 77, 86, 100)
 
# create plot
fig = plt.figure(figsize = (14, 7))
index = np.arange(6)
bar_width = 0.75
opacity = 0.8
 
rects1 = plt.bar(index, Energy, bar_width,
                 alpha=opacity,
                 color=('r', 'orange','orange','y','y','g'))
 
plt.ylabel('Energy(%)', fontsize = 20)
plt.title('Percentage of total energy consumption from wind for various plant configurations', fontsize = 20)
plt.xticks(index, Groups, fontsize = 12)
plt.yticks(fontsize = 15)
plt.legend()
 
plt.tight_layout()
plt.show()
```
